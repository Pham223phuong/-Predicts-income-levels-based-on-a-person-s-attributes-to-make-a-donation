### Handling Outliers

```python
for i in Numerical:
    plt.figure()
    plt.tight_layout()
    
    plt.gca().set(xlabel= i,ylabel='Frequency')
    plt.boxplot(data[i]) 
```

```python
number_of_outliers=[None]*len(Numerical)
q75=[None]*len(Numerical)
q25=[None]*len(Numerical)
IQR=[None]*len(Numerical)
unique_values=[None]*len(Numerical)
ratio=[None]*len(Numerical)

for i, p in enumerate(Numerical):
    q75[i], q25[i]= np.percentile(data[p],[95,5])
    IQR[i]= q75[i]- q25[i]
    #number_of_outliers[i]=df[~((df[p] < (q75[i] + 1.5 * IQR[i])) & (df[p] > (q25[i]-1.5*IQR[i])))].shape[0]
    number_of_outliers[i]=len((np.where((data[p] > (q75[i] + 1.5 * IQR[i])) | (data[p]< (q25[i] - 1.5 * IQR[i])))[0]))
    unique_values[i]= len(data[p].unique())
    ratio[i]= number_of_outliers[i]/unique_values[i]
    print('outliers of',p," =", number_of_outliers[i])
    print('unique_values',p," =", unique_values[i])
    print('ratio',p," =", ratio[i])
    print("*"*40)
```

```python
#Find the outliers records which fall with +/- 5% boundaries 

Q1 = data.quantile(0.05)
Q3 = data.quantile(0.95)
IQR = Q3 - Q1

print("Number of outliers for entire dataset = ", 
    len(data[((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]))
```

It is inevitable to have outliers in the data set. Dropping outliers can make the model less reliable. So, I decided that I will keep them in the dataset.

### Deal with imbalance label

As we have seen above, the 2 labels are very different, so here we will choose proper evaluation metric and Random Undersampling to solve this problem.

The accuracy of a classifier is the total number of correct predictions by the classifier divided by the total number of predictions. This may be good enough for a well-balanced class but not ideal for the imbalanced class problem. The other metrics such as precision is the measure of how accurate the classifier’s prediction of a specific class and recall is the measure of the classifier’s ability to identify a class.

For an imbalanced class dataset F1 score is a more appropriate metric. It is the harmonic mean of precision and recall.

### Model

```python
# Import the three supervised learning models from sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
```

### Decision Tree Classifier

```python
dt = DecisionTreeClassifier(criterion = 'entropy', random_state = 20)
dt.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay, classification_report

y_pred = dt.predict(X_test)
cm = confusion_matrix(y_test, y_pred, labels=dt.classes_)
print("confusion_matrix:\n", cm)
print("accuracy_score:", accuracy_score(y_test, y_pred).round(4))
```

```python
print(classification_report(y_test,y_pred))
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=dt.classes_)
disp.plot(cmap='Blues')
print("      Decision Tree")
```

## 6. Results

- AdaBoost is the most appropriate for our task.

- It performs the best on the testing data, in terms of both the accuracy and f-score.

- SVM is the next best classifier to train on the full training set. So it should scale well even if we have more data.

- By default, Adaboost uses a decision stump i.e. a decision tree of depth 1 as its base classifier, which can handle categorical and numerical data. Weak learners are relatively faster to train, so the dataset size is not a problem for the algorithm.
